```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier

# Split the data into features and target
X = X_df.drop('label', axis=1)
X_df['label'] = y
y = X_df['label']

# Set the number of folds and initialize a new column to store the fold number
num_folds = 10
X['fold'] = -1

# Create a KFold object and use it to assign fold numbers to each row
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)
for fold, (train_index, test_index) in enumerate(kf.split(X)):
    X.loc[test_index, 'fold'] = fold
```

```python
# Initialize a list to store the model's scores
scores = []

# Loop through each fold and fit a linear regression model
for fold in range(num_folds):
    # Get the training and test sets for this fold
    X_train = X[X['fold'] != fold].drop('fold', axis=1)
    y_train = y[X['fold'] != fold]
    X_test = X[X['fold'] == fold].drop('fold', axis=1)
    y_test = y[X['fold'] == fold]
    
    # Fit a linear regression model to the training data
    model = DecisionTreeClassifier(max_depth=2)
    model.fit(X_train, y_train)
    
    # Evaluate the model on the test data
    score = cross_val_score(model, X_test, y_test, cv=kf)
    # Append the score to the list of scores
    scores.append(score)

# Compute the average score across all folds
avg_score = sum(scores) / len(scores)
print(f'Average score: {avg_score}');
#average can changes all time???
```

```python
# Perform cross validation and compute the accuracy for each fold
fold_accuracies = cross_val_score(model, X, y, cv=kf)

# Compute the mean accuracy across all folds
mean_accuracy = fold_accuracies.mean()

# Print the accuracy for each fold and the mean accuracy
for fold, accuracy in enumerate(fold_accuracies):
    print(f'Fold {fold + 1} accuracy: {accuracy:.3f}')
print(f'Mean accuracy: {mean_accuracy:.3f}')
```

##เพื่อนทำถูก90%

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
# Load the iris dataset (โหลดชุดข้อมูล iris)
iris = load_iris()
X = iris.data #แยกข้อมูล iris.data เป็น Input โดยใช้เป็นตัวแปร X
y = iris.target #แยกข้อมูล iris.target เป็น Output โดยใช้เป็นตัวแปร y

# Create a decision tree classifier (สร้างต้นไม้ตัดสินใจ)
clf = DecisionTreeClassifier()

# Create a 10-fold cross-validation object (สร้างการวนลูป 10 fold)
kf = KFold(n_splits=10) 

# Loop through each fold and train/evaluate the decision tree วนซ้ำแต่ละส่วนและประเมินแผนผังการตัดสินใจ
for train_index, test_index in kf.split(X):
    # Get the training and testing data for this fold #รับข้อมูล training and testing 
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Train the decision tree on the training data for this fold ตัดสินใจเกี่ยวกับข้อมูล training 
    clf.fit(X_train, y_train)
    
    # Evaluate the decision tree on the testing data for this fold ประเมินแผนผังการตัดสินใจของ training data สำหรับ  fold
    score = clf.score(X_test, y_test)
    
    # Print the accuracy for this fold #แสดงค่าความแม่นยำ
    print("Accuracy: %0.2f" % score)
```

#add Precision & Recall from ChatGPT

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import precision_score, recall_score

# Load the iris dataset
iris = load_iris()

# Define the decision tree model
model = DecisionTreeClassifier()

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
precision_scores = []
recall_scores = []
for train_idx, test_idx in kf.split(iris.data, iris.target):
    # Get the training and testing data for this fold
    X_train, y_train = iris.data[train_idx], iris.target[train_idx]
    X_test, y_test = iris.data[test_idx], iris.target[test_idx]

    # Train the model on the training data for this fold
    model.fit(X_train, y_train)

    # Predict the labels for the testing data for this fold
    y_pred = model.predict(X_test)

    # Calculate the precision and recall for this fold
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')

    # Add the precision and recall scores to the lists for all folds
    precision_scores.append(precision)
    recall_scores.append(recall)

# Print the average precision and recall scores across all folds
print("Average precision score: %0.2f" % (sum(precision_scores) / len(precision_scores)))
print("Average recall score: %0.2f" % (sum(recall_scores) / len(recall_scores)))
```
